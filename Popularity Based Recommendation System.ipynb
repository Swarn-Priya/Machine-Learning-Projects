# Popularity Based Recommendation System

```python
rating_with_name = ratings.merge(books, on='ISBN')
rating_with_name.head(1)
```

```python
num_rating = rating_with_name.groupby("Book-Title").count()['Book-Rating'].reset_index()
num_rating.rename(columns={'Book-Rating':"Num of Ratings"},inplace=True)
num_rating
```

```python
avg_rating = rating_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating.rename(columns={'Book-Rating' : 'Averge Rating'},inplace=True)
avg_rating
```

### Popular Book Table

```python
popular_df = num_rating.merge(avg_rating,on='Book-Title')
popular_df
```

```python
popular_df = popular_df[popular_df['Num of Ratings']>=250].sort_values("Averge Rating",ascending=False).head(50)

#Top 50 books which has got good number of rating and average rating is also high. 
```
