```python
# Spliting the Data

x = df.iloc[:,0:-1]
y = df['Class']
```

```python
# Feature SCaling

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
scx = sc.fit_transform(x)
scx
```

```python
# Train and Test SPlit

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state = 101)
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

### MOdel BUilding

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier

from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
Mlist = {'LogisticRegression':LogisticRegression(),
         'DecisionTreeClassifier':DecisionTreeClassifier(criterion='gini'),
         'RandomForestClassifier':RandomForestClassifier(n_estimators=100),
#          'KNeighborsClassifier':KNeighborsClassifier(n_neighbors=5),
#          'SVC':SVC(),
         'AdaBoostClassifier':AdaBoostClassifier(learning_rate=0.001),
#          'GradientBoostingClassifier': GradientBoostingClassifier(learning_rate=0.001),
#          'XGBClassifier' : XGBClassifier(learning_rate=0.001)
        }
```

```python
# TO Calculate the best n_neighbor size...but taking too much time. Sonot executing it at this momnet.

# error_rate = []

# for i in range(2,15):
#     knn =  KNeighborsClassifier(n_neighbors=i)
#     knn.fit(x_train,y_train)
#     y_pred=knn.predict(x_test)
#     error_rate.append(np.mean(y_pred != y_test))
```

```python

for name,modl in Mlist.items():
    name=modl
    name.fit(x_train,y_train)
    y_pred_test=name.predict(x_test)
    y_pred_train = name.predict(x_train)
    print("*"*3,name,"*"*3)
    #Accurcy Score
    print("Accuracy Score Train - ",accuracy_score(y_train,y_pred_train))
    print("Accuracy Score Test - ",accuracy_score(y_test,y_pred_test))
    print(classification_report(y_train,y_pred_train))
    print(classification_report(y_test,y_pred_test))
    print(confusion_matrix(y_train,y_pred_train))
    print(confusion_matrix(y_test,y_pred_test))
    print("________________________"*3)
```

# BALANCING DATASET

```python
df['Class'].value_counts()
```

```python
import imblearn
from imblearn.over_sampling import SMOTE

smote = SMOTE()
x_resample,y_resample = smote.fit_resample(x,y)
```

```python
print(x_resample.shape,y_resample.shape)
print(x.shape,y.shape)
```

```python
scx = sc.fit_transform(x_resample)
# scx
xr_train,xr_test,yr_train,yr_test = train_test_split(x_resample,y_resample,test_size=0.2,random_state = 101)
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
print("Resampled data - ",xr_train.shape,xr_test.shape,yr_train.shape,yr_test.shape)
```

```python
for name,modl in Mlist.items():
    name=modl
    name.fit(xr_train,yr_train)
    y_pred_test=name.predict(xr_test)
    y_pred_train = name.predict(xr_train)
    print("*"*3,name,"*"*3)
    #Accurcy Score
    print("Accuracy Score Train - ",accuracy_score(yr_train,y_pred_train))
    print("Accuracy Score Test - ",accuracy_score(yr_test,y_pred_test))
    print(classification_report(yr_train,y_pred_train))
    print(classification_report(yr_test,y_pred_test))
    print(confusion_matrix(yr_train,y_pred_train))
    print(confusion_matrix(yr_test,y_pred_test))
    print("________________________"*3)
```
