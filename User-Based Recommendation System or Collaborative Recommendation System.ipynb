# Collaborative Recommendation System

```python
rating_with_name = ratings.merge(books,on='ISBN')
rating_with_name.head(1)
```

```python
rating_with_name.shape
```

```python
x = rating_with_name.groupby('User-ID').count()['Book-Rating']>200
educated_user = x[x].index
```

```python
educated_user
```

```python
filtered_rating = rating_with_name[rating_with_name['User-ID'].isin(educated_user)]
filtered_rating
```

```python
filtered_rating.shape
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index  ## WIll get the book title. 
```

```python
famous_books
```

```python
final_rating=filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
final_rating
```

# Algorythm

```python
pt=final_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
pt.shape
# pt
```

```python
pt.fillna(0,inplace=True)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
def recommendation_system(book_name):
#     print(book_name)
    #index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:6]
    data=[]
    
    for i in similar_items:
        item =[]
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-L'].values))
        
        data.append(item)
        
    return data
```

```python
recommendation_system('Harry Potter and the Prisoner of Azkaban (Book 3)')
```
