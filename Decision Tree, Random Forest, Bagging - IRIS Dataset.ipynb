# EDA COMPLETE

# Spliting the data

```python
x = iris.drop(['species'],axis=1)
y=iris['species']
```

# Feature Selection

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ssx = ss.fit_transform(x)
ssx
```

# Training and testing split.

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(ssx,y,test_size=0.2,random_state = 217)
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

# Model Building

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier(criterion='gini')
dtree.fit(x_train,y_train)

dtree1 = DecisionTreeClassifier(criterion='entropy')
dtree1.fit(x_train,y_train)
```

```python
y_pred_dt_train = dtree.predict(x_train)
y_pred_dt_test = dtree.predict(x_test)

y_pred_dt_train1 = dtree1.predict(x_train)
y_pred_dt_test1 = dtree1.predict(x_test)
```

# Decision Tree Evaluation

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
print("Accuracy SCore - Training", accuracy_score(y_train,y_pred_dt_train))
print("***********************")
print("Accuracy SCore - Testing", accuracy_score(y_test,y_pred_dt_test))
print()
print()
print("Accuracy SCore ENtropy - Training", accuracy_score(y_train,y_pred_dt_train1))
print("***********************")
print("Accuracy SCore ENtropy - Testing", accuracy_score(y_test,y_pred_dt_test1))
```

```python
print(classification_report(y_train,y_pred_dt_train))
print("*************************************************************")
print(classification_report(y_test,y_pred_dt_test))
print()
print()
print()
print()
print(classification_report(y_train,y_pred_dt_train1))
print("*************************************************************")
print(classification_report(y_test,y_pred_dt_test1))
```

```python
confusion_matrix(y_test,y_pred_dt_test)
```

```python
confusion_matrix(y_train,y_pred_dt_train)
```

```python
# User Defined Function

def report_model(model):
    model_preds = model.predict(x_test)
    print(classification_report(y_test,model_preds))
    print('\n')
    
    plt.figure(figsize=(8,12))
    plot_tree(model,feature_names = x.columns,filled = True)
```

# Bagging & RandomForestClassifier

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import RandomForestClassifier
```

1. Bagging

```python
bag = BaggingClassifier()
bag.fit(x_train,y_train)
```

```python
y_pred_bag_train = bag.predict(x_train)
y_pred_bag_test = bag.predict(x_test)
```

```python
print("Accuracy SCore Bagging - Training", accuracy_score(y_train,y_pred_bag_train))
print("***********************")
print("Accuracy SCore Bagging - Testing", accuracy_score(y_test,y_pred_bag_test))
```

2. Random Forest

```python
rf = RandomForestClassifier(n_estimators=100,criterion='gini')
rf.fit(x_train,y_train)
```

```python
y_pred_rf_train = rf.predict(x_train)
y_pred_rf_test = rf.predict(x_test)
```

```python
rf = RandomForestClassifier(n_estimators=500,criterion='entropy')
rf.fit(x_train,y_train)
```

```python
y_pred_rfen_train = rf.predict(x_train)
y_pred_rfen_test = rf.predict(x_test)
```

# Evaluating Random FOrest

```python
print("Accuracy SCore - Training - ENtropy", accuracy_score(y_train,y_pred_rfen_train))
print("***********************")
print("Accuracy SCore - Testing Entropy ", accuracy_score(y_test,y_pred_rfen_test))
```

```python
print("Accuracy SCore - Training -  Gini", accuracy_score(y_train,y_pred_rf_train))
print("***********************")
print("Accuracy SCore - Testing - Gini", accuracy_score(y_test,y_pred_rf_test))
```

```python
print("Entropy")
print(classification_report(y_train,y_pred_rfen_train))
print(classification_report(y_test,y_pred_rfen_test))

print("Gini")
print(classification_report(y_train,y_pred_rf_train))
print(classification_report(y_test,y_pred_rf_test))
```

```python
print("Entropy")
print(confusion_matrix(y_train,y_pred_rfen_train))
print(confusion_matrix(y_test,y_pred_rfen_test))
print()
print("Gini")
print(confusion_matrix(y_train,y_pred_rf_train))
print(confusion_matrix(y_test,y_pred_rf_test))
```
