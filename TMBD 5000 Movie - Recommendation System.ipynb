# AST - ABstract Syntax Tree

It is used to extract information from the given mixed format. 

```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i["name"])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head(1)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
def convert3(text):
    L=[]
    counter = 0
    for i in ast.literal_eval(text):
        if counter <3:
            L.append(i['name'])
        counter += 1
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
def convertdir(text):
    L=[]
    for i in ast.literal_eval(text):
        if i['job'] =='Director':
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(convertdir)
```

```python
movies.head()
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head(1)
```

```python
movies.dropna(inplace = True)
```

```python
new = movies.drop(columns=['overview','cast','crew','genres','keywords'],axis = 1)
```

# NLP Concept

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(stop_words ='english',max_features = 5000,binary = True)
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
cv.get_feature_names()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity.shape
```

```python
def recommned(movie):
    index = new[new['title'] == movies].index[0]
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distance = sorted(list(enumerate(similarity[index])) , reverse= True , key = lambda x:x[1])
    for i in distance[1:6]:
        print(new.iloc[i[0]].title)
```

```python
recommend('Titanic')
```
